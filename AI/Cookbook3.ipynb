```python
import pandas as pd
import numpy as np
import csv
```

```python
zeros = np.zeros((7,7))
zeros
```

```python
df = pd.DataFrame(zeros, columns=['A', 'B', 'C', 'D', 'E', 'F', 'G'], index=['A', 'B', 'C', 'D', 'E', 'F', 'G'])
df
```

```python
with open ("Cookbook3.csv") as csvfile:
    readCsv = csv.reader(csvfile, delimiter=',')
    for row in readCsv:
        df.at[row[0], row[1]] = row[2]
        df.at[row[1], row[0]] = row[2]
df
```

```python
# iloc allows you to access data with positions of the row or column indexes
# iloc[:,3] gets the fourth row of the dataframe despite the label being 'D'
iloc = df.iloc[3,:]
iloc
```

```python
# loc allows you to access data with assigned index labels for row or columns
# loc[:,D] gets the sixth row of the dataframe
loc = df.loc[:,'F']
loc
```

```python
# Slicing example
# slicer gets the last row of the table
slicer = df[-1:]
slicer
```

```python
# Compute the averages of the rows
max_average = max(df.mean(axis=1))
higher_average = []
sums = df.sum(axis=1)
for row in range(len(df.index)):
    if sums[row] > max_average:
        print("Row", row, "has a sum of", sums[row], "which is greater than the max average of", max_average)
```

```python
row_d = df.loc['D',:]
row_d
squared_row_d_for = []
for i in row_d:
    squared_row_d_for.append(i**2)
print("Squared row D using for loops", squared_row_d_for)
squared_row_d_list_comp = [i**2 for i in row_d]
print("Squared row D using list comprehension", squared_row_d_list_comp)
```
